```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException, TimeoutException
from selenium.webdriver.common.keys import Keys
import time

# ChromeDriver 옵션 설정
options = webdriver.ChromeOptions()
options.add_argument("--disable-gpu")
options.add_argument("--window-size=1920,1080")
options.add_experimental_option("excludeSwitches", ["enable-logging", "enable-automation"])
options.add_argument("--disable-blink-features=AutomationControlled")

# ChromeDriver 초기화
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=options)

# 카카오맵 웹사이트로 이동
driver.get('https://map.kakao.com/')

# 페이지 로드 대기
time.sleep(5)

# 검색창 찾기 및 검색어 입력
search_box = driver.find_element(By.ID, "search.keyword.query")
search_box.send_keys("뱅뱅막국수")  # 원하는 검색어 입력
search_box.send_keys(Keys.ENTER)  # 엔터 키로 검색 실행

# 검색 결과 로딩 대기
time.sleep(2)

try:
    # JavaScript를 사용하여 강제로 클릭
    first_detail_link = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="info.search.place.list"]/li[1]/div[5]/div[4]/a[1]'))
    )
    driver.execute_script("arguments[0].click();", first_detail_link)
    print("상세보기 클릭 완료")
except (ElementClickInterceptedException, TimeoutException) as e:
    print("상세보기를 클릭할 수 없습니다:", e)

# 페이지가 로드될 시간을 줌
time.sleep(3)

# 새 창으로 포커스를 변경
original_window = driver.current_window_handle
for window_handle in driver.window_handles:
    if window_handle != original_window:
        driver.switch_to.window(window_handle)
        break

# 리뷰 요소 찾기
all_reviews = []
previous_review_count = 0  # 이전 리뷰 수

# 스크롤을 반복하여 리뷰를 가져옴
while True:
    # 리뷰 요소를 찾기
    try:
        # 모든 리뷰를 가져오기
        review_elements = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.XPATH, '//*[@id="mArticle"]/div[8]/div[3]/ul/li'))
        )

        for review in review_elements:
            try:
                # 리뷰 텍스트 가져오기 (각 리뷰마다 다를 수 있는 XPath)
                review_text = review.find_element(By.XPATH, './/p/span').text  # p/span 경로가 다를 수 있으니 유동적으로 가져옴
                if review_text not in all_reviews:  # 중복 제거
                    all_reviews.append(review_text)
            except NoSuchElementException:
                continue  # 리뷰 요소가 없을 경우 무시

        # 스크롤을 내림
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)  # 스크롤 후 대기

        # 새로운 리뷰 수를 확인
        if len(all_reviews) == previous_review_count:
            print("더 이상 새로운 리뷰가 없습니다.")
            break  # 더 이상 새로운 리뷰가 없으면 종료
        previous_review_count = len(all_reviews)  # 현재 리뷰 수를 업데이트

    except NoSuchElementException:
        print("리뷰 요소를 찾을 수 없습니다.")
        break
    except TimeoutException:
        print("리뷰 요소 로드가 시간 내에 완료되지 않았습니다.")
        break

# 모든 리뷰 출력
if all_reviews:
    for i, review in enumerate(all_reviews):
        print(f"리뷰 {i + 1}: {review}")
else:
    print("리뷰가 없습니다.")

# 드라이버 종료 (필요 시)
# driver.quit()
```
