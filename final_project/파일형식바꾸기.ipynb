```python
import pandas as pd
from dateutil import parser
import re
import os

# 경로 설정
input_folder = "C:/fintech_service/final_project/data/리뷰추출/네이버수정전"
output_folder = "C:/fintech_service/final_project/data/리뷰추출/네이버"
os.makedirs(output_folder, exist_ok=True)

# 날짜 형식을 변환하는 함수
def convert_date(date_str):
    # '2023년 9월 1일 금요일' 형식을 '2023년 9월 1일'로 변경하기 위한 정규식
    date_str = re.sub(r'(\d{4}년 \d{1,2}월 \d{1,2}일) \w+', r'\1', date_str)
    
    # 다양한 형식을 명확하게 시도하여 변환
    formats = [
        "%Y년 %m월 %d일",       # 예시: 2023년 9월 1일
        "%Y/%m/%d",             # 예시: 2023/09/01
        "%Y-%m-%d",             # 예시: 2023-09-01
        "%Y.%m.%d",             # 예시: 2023.09.01
        "%Y%m%d"                # 예시: 20230901
    ]
    
    for fmt in formats:
        try:
            return pd.to_datetime(date_str, format=fmt).strftime('%Y-%m')
        except ValueError:
            continue
    
    # dateutil의 parser로 마지막 시도
    try:
        # 문자열 자동 감지로 변환
        return parser.parse(date_str).strftime('%Y-%m')
    except (ValueError, TypeError):
        # 모든 시도가 실패하면 예외 발생
        raise ValueError(f"Unsupported date format: {date_str}")

# 입력 폴더에서 모든 CSV 파일 처리
for filename in os.listdir(input_folder):
    if filename.endswith(".csv"):
        input_file = os.path.join(input_folder, filename)

        # CSV 파일 불러오기
        df = pd.read_csv(input_file)

        # 'Index' 열 삭제
        if 'Index' in df.columns:
            df = df.drop(columns=['Index'])

        # 'Date' 열이 있는 경우에만 변환 시도
        if 'Date' in df.columns and df['Date'].dtype == 'object':
            try:
                df['Date'] = df['Date'].apply(convert_date)
            except ValueError as e:
                print(f"파일 처리 중 오류 발생: {filename} - {e}")
                continue

        # 변환된 데이터를 새로운 파일로 저장
        output_file = os.path.join(output_folder, filename)
        df.to_csv(output_file, index=False, encoding='utf-8-sig')
        print(f"{filename} 처리 및 저장 완료.")

print("모든 파일이 변환되어 저장되었습니다.")
```

# 카카오 형식 변경 및 더보기 제거

```python
import pandas as pd
from dateutil import parser
import re
import os

# 경로 설정
input_folder = "C:/fintech_service/final_project/data/리뷰추출/카카오수정전"
output_folder = "C:/fintech_service/final_project/data/리뷰추출/카카오"
os.makedirs(output_folder, exist_ok=True)

# 날짜 형식을 변환하는 함수
def convert_date(date_str):
    # '2023년 9월 1일 금요일' 형식을 '2023년 9월 1일'로 변경
    date_str = re.sub(r'(\d{4}년 \d{1,2}월 \d{1,2}일) \w+', r'\1', date_str)
    
    # 다양한 형식을 시도하여 변환
    formats = [
        "%Y년 %m월 %d일",       # 예시: 2023년 9월 1일
        "%Y/%m/%d",             # 예시: 2023/09/01
        "%Y-%m-%d",             # 예시: 2023-09-01
        "%Y.%m.%d",             # 예시: 2023.09.01
        "%Y%m%d"                # 예시: 20230901
    ]
    
    for fmt in formats:
        try:
            return pd.to_datetime(date_str, format=fmt).strftime('%Y-%m')
        except ValueError:
            continue
    
    # dateutil의 parser로 마지막 시도
    try:
        return parser.parse(date_str).strftime('%Y-%m')
    except (ValueError, TypeError):
        raise ValueError(f"Unsupported date format: {date_str}")

# 리뷰 내용에서 '더보기' 제거 함수
def clean_review(text):
    if isinstance(text, str):
        return text.replace("더보기", "").strip()
    return text

# 폴더의 모든 CSV 파일 처리
for file_name in os.listdir(input_folder):
    if file_name.endswith(".csv"):
        input_file = os.path.join(input_folder, file_name)
        
        # CSV 파일 읽기
        df = pd.read_csv(input_file)
        
        # 컬럼명 변경
        df.rename(columns={
            '작성 날짜': 'Date',
            '리뷰 내용': 'Review'
        }, inplace=True)
        
        # 'Date' 열 처리
        if 'Date' in df.columns and df['Date'].dtype == 'object':
            df['Date'] = df['Date'].apply(convert_date)
        
        # 'Review' 열에서 '더보기' 제거
        if 'Review' in df.columns:
            df['Review'] = df['Review'].apply(clean_review)
        
        # '업종 정보' 열 삭제 (존재하는 경우)
        if '업종 정보' in df.columns:
            df = df.drop(columns=['업종 정보'])
        
        # 저장 경로 설정
        output_file = os.path.join(output_folder, file_name)
        
        # 변환된 데이터 저장
        df.to_csv(output_file, index=False, encoding='utf-8-sig')

print("모든 카카오 리뷰 파일이 변환되고 저장되었습니다.")
```

# 구글

```python
import pandas as pd
from datetime import datetime, timedelta
import re
import os

# 현재 날짜 설정
current_date = datetime(2024, 11, 1)  # 기준 날짜를 2024년 11월 1일로 설정

# 입력 및 출력 폴더 경로 설정
input_folder = r'C:\fintech_service\final_project\data\리뷰추출\구글수정전'
output_folder = r'C:\fintech_service\final_project\data\리뷰추출\구글'

# 날짜 형식을 변환하는 함수
def convert_google_date(date_str):
    # "n년 전" 처리
    match_year = re.match(r'(\d+)년 전', date_str)
    if match_year:
        years_ago = int(match_year.group(1))
        target_date = current_date.replace(year=current_date.year - years_ago, month=1, day=1)
        return target_date.strftime('%Y-%m')
    
    # "n달 전" 처리
    match_month = re.match(r'(\d+)달 전', date_str)
    if match_month:
        months_ago = int(match_month.group(1))
        target_date = current_date - timedelta(days=30 * months_ago)
        return target_date.strftime('%Y-%m')
    
    # "n주 전" 처리
    match_week = re.match(r'(\d+)주 전', date_str)
    if match_week:
        weeks_ago = int(match_week.group(1))
        target_date = current_date - timedelta(weeks=weeks_ago)
        return target_date.strftime('%Y-%m')

    # "n일 전" 처리
    match_day = re.match(r'(\d+)일 전', date_str)
    if match_day:
        days_ago = int(match_day.group(1))
        target_date = current_date - timedelta(days=days_ago)
        return target_date.strftime('%Y-%m')

    # 기타 형식은 기본 반환 (변환 불가 시 예외 처리)
    raise ValueError(f"Unsupported date format: {date_str}")

# 폴더 내 모든 CSV 파일 순회
for file_name in os.listdir(input_folder):
    # 파일이 CSV인 경우만 처리
    if file_name.endswith('.csv'):
        input_file_path = os.path.join(input_folder, file_name)
        output_file_path = os.path.join(output_folder, file_name)

        # CSV 파일 불러오기
        df = pd.read_csv(input_file_path)

        # 'Date' 열 변환
        if 'Date' in df.columns and df['Date'].dtype == 'object':
            df['Date'] = df['Date'].apply(convert_google_date)

        # 변환된 데이터를 출력 폴더에 저장
        df.to_csv(output_file_path, index=False, encoding='utf-8-sig')

print("모든 파일의 변환이 완료되었습니다!")
```
