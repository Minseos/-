```python
categoryMapping = {
    '한식': ['한식', '곰탕,설렁탕', '돼지고기구이', '낙지요리', '백반,가정식', '육류,고기요리', '순대,순댓국', '찌개,전골', '정육식당', '장어,먹장어요리', '족발,보쌈', '한식뷔페', '닭요리', '국수'],
    '중식': ['중식당'],
    '일식': ['일식당', '생선회', '초밥,롤', '돈가스'],
    '양식': ['이탈리아음식', '햄버거'],
    '카페/디저트': ['카페', '카페,디저트', '베이커리'],

    '분식': ['분식', '종합분식', '김밥'],
    '치킨': ['치킨,닭강정'],
    '해물': ['게요리', '매운탕,해물탕', '해물,생선요리'],
    '요리주점': ['요리주점', '바(BAR)', '맥주,호프', '포장마차'],
    '다이어트': ['샌드위치']
}
```

```python
# 플랫폼별 리뷰 개수 계산
platform_mapping = {'네이버': 'naver_count', '카카오': 'kakao_count', '구글': 'google_count'}
review_data['platform'] = review_data['platform'].map(platform_mapping)  # 플랫폼 이름 매핑

# 리뷰 개수를 계산
platform_counts = review_data.groupby(['store_id', 'platform']).size().unstack(fill_value=0)

# store_data에 플랫폼별 리뷰 개수 추가
for platform in platform_mapping.values():
    if platform in platform_counts.columns:
        store_data[platform] = store_data['store_id'].map(platform_counts[platform])
    else:
        store_data[platform] = 0  # 해당 플랫폼이 없으면 0으로 채움

# 결과 확인
store_data.head()
```

```python
# 가게별 총 리뷰 개수 계산
store_data['total_count'] = store_data[['naver_count', 'kakao_count', 'google_count']].sum(axis=1)

# 결과 확인
store_data[['store_id', 'store_name', 'naver_count', 'kakao_count', 'google_count', 'total_count']].head()
```

```python
import numpy as np
import pandas as pd

# 신뢰도 값 정의
trust_scores = {'네이버': 3.30, '카카오': 3.58, '구글': 3.64}

# 가중 평점(weighted_rating) 계산 함수
def calculate_weighted_rating(row):
    platforms = ['naver', 'kakao', 'google']
    total_weight = 0
    weighted_sum = 0

    for platform in platforms:
        # 한글 키로 매핑
        platform_kor = {'naver': '네이버', 'kakao': '카카오', 'google': '구글'}[platform]
        review_count = row[f'{platform}_count']
        rating = row[f'{platform}_rating']
        trust_score = trust_scores[platform_kor]  # 신뢰도 값 가져오기

        # 유효한 데이터만 포함 (0 이상의 평점만 고려)
        if pd.notna(review_count) and review_count > 0 and pd.notna(rating) and rating > 0:
            weight = review_count * trust_score  # 가중치 계산
            total_weight += weight
            weighted_sum += rating * weight  # 가중 합계 계산

    # 총 가중치가 0인 경우 0 반환
    return weighted_sum / total_weight if total_weight > 0 else 0

# 가중 평점 계산
store_data['weighted_rating'] = store_data.apply(calculate_weighted_rating, axis=1)

# 결과 확인
result = store_data[['store_name', 'naver_rating', 'kakao_rating', 'google_rating', 'weighted_rating']]
print(result)
```

```python
import pandas as pd

# 긍정/부정 비율 계산 함수
def calculate_ratios(group, column):
    counts = group[column].value_counts().to_dict()
    positive = counts.get('긍정적', 0)
    negative = counts.get('부정적', 0)
    total = positive + negative  # "알 수 없음"은 제외
    return {
        f'final_{column}_positive': positive / total if total > 0 else 0,
        f'final_{column}_negative': negative / total if total > 0 else 0,
    }

# 가게별로 데이터 그룹화 및 계산
final_results = []
grouped_reviews = review_data.groupby('store_id')

for store_id, group in grouped_reviews:
    # store_data에서 store_name 가져오기
    store_name = store_data.loc[store_data['store_id'] == store_id, 'store_name'].values[0] \
        if store_id in store_data['store_id'].values else f"Store_{store_id}"
    
    # 각 항목별 비율 계산
    taste_ratios = calculate_ratios(group, 'taste')
    service_ratios = calculate_ratios(group, 'service')
    quantity_ratios = calculate_ratios(group, 'quantity')
    
    # 결과 저장
    final_results.append({
        'store_id': store_id,
        'store_name': store_name,
        **taste_ratios,
        **service_ratios,
        **quantity_ratios,
    })

# 최종 데이터프레임 생성
final_results_df = pd.DataFrame(final_results)

# 결과 확인
print(final_results_df)

# 결과를 CSV로 저장
output_file_path = './store_sentiment_ratios.csv'
final_results_df.to_csv(output_file_path, index=False, encoding='utf-8-sig')
print(f"분석 결과가 {output_file_path} 파일로 저장되었습니다.")
```
