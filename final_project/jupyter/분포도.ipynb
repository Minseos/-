```python
import pandas as pd

# CSV 파일 로드
store_data = pd.read_csv('../data/store_data.csv')

# 카테고리 매핑
category_mapping = {
    '한식': ['한식', '곰탕,설렁탕', '돼지고기구이', '낙지요리', '백반,가정식', '육류,고기요리', '순대,순댓국', '찌개,전골', '정육식당', '장어,먹장어요리', '족발,보쌈', '한식뷔페', '닭요리', '국수'],
    '중식': ['중식당'],
    '일식': ['일식당', '생선회', '초밥,롤', '돈가스'],
    '양식': ['이탈리아음식', '햄버거'],
    '카페/디저트': ['카페', '카페,디저트', '베이커리']
}

# 카테고리별 store_id 추출
store_ids_by_category = {}

for category, keywords in category_mapping.items():
    filtered_df = store_data[store_data['category'].isin(keywords)]  # 키워드 필터링
    store_ids = filtered_df['store_id'].tolist()  # store_id 추출
    if store_ids:
        store_ids_by_category[category] = store_ids

# 결과 확인
print(store_ids_by_category)
```

```python
import shutil
import os

# 카테고리와 관련 이미지 맵핑
category_mapping = {
    '한식': ['한식', '곰탕,설렁탕', '돼지고기구이', '낙지요리', '백반,가정식', '육류,고기요리', '순대,순댓국', '찌개,전골', '정육식당', '장어,먹장어요리', '족발,보쌈', '한식뷔페', '닭요리', '국수'],
    '중식': ['중식당'],
    '일식': ['일식당', '생선회', '초밥,롤', '돈가스'],
    '양식': ['이탈리아음식', '햄버거'],
    '카페/디저트': ['카페', '카페,디저트', '베이커리'],
}

# 이미지 매핑
image_mapping = {
    "한식": "korean_weighted_rating_vs_price_range_actual_values.png",
    "카페/디저트": "cafe_weighted_rating_vs_price_range.png",
    "중식": "chinese_weighted_rating_vs_price_range.png",
    "일식": "japanese_weighted_rating_vs_price_range.png",
    "양식": "western_weighted_rating_vs_price_range.png"
}

# 가게 ID 목록 (예시: 중식 카테고리)
store_ids_by_category = {
    '한식': [2, 6, 7, 11, 13, 20, 21, 32, 35, 36, 38, 39, 47, 49, 58, 62, 63, 66, 68, 71, 73, 74, 75, 77, 79, 80, 89, 94, 95],
    '중식': [26, 40, 44, 78],
    '일식': [8, 27, 41, 50, 57, 69, 76],
    '양식': [14, 25, 28],
    '카페/디저트': [5, 17, 24, 34, 42, 51, 54, 55, 56, 59, 60, 64, 65, 83, 85, 90, 91, 92, 96]
}

# 경로 설정
base_image_dir = "C:/study/fintech/final_project/data/사진/3_분포도"  # 이미지 폴더 경로
output_dir = "C:/study/fintech/final_project/data/사진/3_분포도"  # 저장 경로
os.makedirs(output_dir, exist_ok=True)

# 카테고리별 이미지 복사 및 이름 변경
for category, store_ids in store_ids_by_category.items():
    image_name = image_mapping.get(category)  # 카테고리별 이미지 이름
    if image_name:
        source_image_path = os.path.join(base_image_dir, image_name)
        
        # 각 store_id에 대해 이미지 복사 및 이름 변경
        for store_id in store_ids:
            target_file_name = f"{store_id}__Distribution.png"
            target_path = os.path.join(output_dir, target_file_name)
            shutil.copy(source_image_path, target_path)
            print(f"이미지 저장됨: {target_path}")
    else:
        print(f"카테고리 '{category}'에 해당하는 이미지가 없습니다.")
```
