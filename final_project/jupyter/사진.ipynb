```python
import os
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from PIL import Image
import requests
from io import BytesIO

# ChromeDriver 옵션 설정
options = webdriver.ChromeOptions()
options.add_argument("--disable-gpu")
options.add_argument("--window-size=1920,1080")
options.add_experimental_option("excludeSwitches", ["enable-logging", "enable-automation"])
options.add_argument("--disable-blink-features=AutomationControlled")

# ChromeDriver 초기화
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=options)

# 네이버 지도 웹사이트로 이동
driver.get('https://map.naver.com/v5/')
time.sleep(3)  # 페이지 로드 대기

# 한식 가게 리스트 (store_id와 store_name)
store_data = [

    {"store_id": 62, "store_name": "예당"},
    {"store_id": 63, "store_name": "오늘은_어떤닭"},
    {"store_id": 66, "store_name": "왕산골"},
    {"store_id": 68, "store_name": "원할매이모네닭한마리"},
    {"store_id": 71, "store_name": "이천쌀밥"},
    {"store_id": 73, "store_name": "장수촌풍천장어"},
    {"store_id": 74, "store_name": "전주피순대"},
    {"store_id": 75, "store_name": "조순금닭도리탕"},
    {"store_id": 77, "store_name": "지우네"},
    {"store_id": 79, "store_name": "참족"},
    {"store_id": 80, "store_name": "청담배짱이"},
    {"store_id": 89, "store_name": "통돼지_두루치기김치찌개"},
    {"store_id": 94, "store_name": "판문점부대찌개_역삼점"},
    {"store_id": 95, "store_name": "한강수"}
]

# 사진 저장 경로
save_dir = r"C:\study\fintech\final_project\data\사진\가게사진"
os.makedirs(save_dir, exist_ok=True)

for store in store_data:
    store_id = store["store_id"]
    store_name = store["store_name"]

    try:
        # 검색창 요소 대기
        search_box = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "input.input_search"))
        )
        
        # 검색창 초기화
        search_box.click()
        search_box.send_keys(Keys.CONTROL, 'a')  # 기존 내용 선택
        search_box.send_keys(Keys.DELETE)  # 선택된 내용 삭제
        time.sleep(1)  # 초기화 대기

        # 새로운 검색어 입력
        search_box.send_keys(store_name)
        search_box.send_keys(Keys.RETURN)

        # 검색 결과 로드 대기
        time.sleep(3)

        # 검색 결과 확인 (검색 결과가 없으면 다음으로 넘어감)
        no_result_message = driver.find_elements(By.CSS_SELECTOR, ".no_result")
        if no_result_message:
            print(f"Store ID {store_id} ({store_name}) 검색 결과 없음.")
            continue  # 다음 가게로 넘어감

        # 검색 결과 iframe으로 전환
        search_iframe = WebDriverWait(driver, 10).until(
            EC.frame_to_be_available_and_switch_to_it((By.ID, "searchIframe"))
        )

        # 첫 번째 검색 결과 클릭
        first_result_xpath = '//*[@id="_pcmap_list_scroll_container"]/ul/li[1]/div[1]'
        first_result = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, first_result_xpath))
        )
        
        # JavaScript 강제 클릭
        driver.execute_script("arguments[0].scrollIntoView(true);", first_result)  # 스크롤로 이동
        time.sleep(1)  # 이동 대기
        driver.execute_script("arguments[0].click();", first_result)  # 강제 클릭
        time.sleep(3)

        # 메인 컨텐츠로 다시 전환
        driver.switch_to.default_content()

        # entryIframe으로 전환
        entry_iframe = WebDriverWait(driver, 10).until(
            EC.frame_to_be_available_and_switch_to_it((By.ID, "entryIframe"))
        )
        time.sleep(3)

        # 대표 사진 또는 동영상 처리
        photo_elements = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".place_thumb img"))
        )

        # 사진 URL 추출
        photo_url = None
        for photo in photo_elements:
            # 부모 요소에 동영상 관련 클래스가 있는지 확인
            parent_element = photo.find_element(By.XPATH, "..")
            is_video = len(parent_element.find_elements(By.CLASS_NAME, "place_thumb_icon_video")) > 0
            if not is_video:  # 동영상이 아니면 사진 URL 가져오기
                photo_url = photo.get_attribute("src")
                break

        if not photo_url:
            print(f"Store ID {store_id} ({store_name}) 사진이 없습니다.")
            continue

        # 사진 저장
        response = requests.get(photo_url)
        image = Image.open(BytesIO(response.content))
        save_path = os.path.join(save_dir, f"{store_id}.png")
        image.save(save_path, format="PNG")  # PNG 형식으로 저장
        print(f"Store ID {store_id} 사진 저장 완료: {save_path}")

        # iframe 전환 초기화
        driver.switch_to.default_content()

    except Exception as e:
        print(f"Store ID {store_id} ({store_name}) 처리 중 오류 발생: {e}")

# 드라이버 종료
driver.quit()
```

```python
import os
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from PIL import Image
import requests
from io import BytesIO

# ChromeDriver 옵션 설정
options = webdriver.ChromeOptions()
options.add_argument("--disable-gpu")
options.add_argument("--window-size=1920,1080")
options.add_experimental_option("excludeSwitches", ["enable-logging", "enable-automation"])
options.add_argument("--disable-blink-features=AutomationControlled")

# ChromeDriver 초기화
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=options)

# 네이버 지도 웹사이트로 이동
driver.get('https://map.naver.com/v5/')
time.sleep(3)  # 페이지 로드 대기

store_data = [
    {"store_id": 26, "store_name": "리첸"},
    {"store_id": 40, "store_name": "북경반점"},
    {"store_id": 44, "store_name": "사천루"},
    {"store_id": 78, "store_name": "짜짜루"}
]


# 사진 저장 경로
save_dir = r"C:\study\fintech\final_project\data\사진\가게사진"
os.makedirs(save_dir, exist_ok=True)

for store in store_data:
    store_id = store["store_id"]
    store_name = store["store_name"]

    try:
        # 검색창 요소 대기
        search_box = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "input.input_search"))
        )
        
        # 검색창 초기화
        search_box.click()
        search_box.send_keys(Keys.CONTROL, 'a')  # 기존 내용 선택
        search_box.send_keys(Keys.DELETE)  # 선택된 내용 삭제
        time.sleep(1)

        # 새로운 검색어 입력
        search_box.send_keys(store_name)
        search_box.send_keys(Keys.RETURN)

        # 검색 결과 로드 대기
        time.sleep(3)

        # 검색 결과 확인 (검색 결과가 없으면 다음으로 넘어감)
        no_result_message = driver.find_elements(By.CSS_SELECTOR, ".no_result")
        if no_result_message:
            print(f"Store ID {store_id} ({store_name}) 검색 결과 없음.")
            continue

        # 검색 결과 iframe으로 전환
        search_iframe = WebDriverWait(driver, 10).until(
            EC.frame_to_be_available_and_switch_to_it((By.ID, "searchIframe"))
        )

        # 첫 번째 검색 결과 클릭
        first_result_xpath = '//*[@id="_pcmap_list_scroll_container"]/ul/li[1]/div[1]'
        first_result = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, first_result_xpath))
        )
        
        # ActionChains로 클릭
        actions = ActionChains(driver)
        actions.move_to_element(first_result).click().perform()
        time.sleep(2)

        # 메인 컨텐츠로 다시 전환
        driver.switch_to.default_content()

        # entryIframe으로 전환
        entry_iframe = WebDriverWait(driver, 10).until(
            EC.frame_to_be_available_and_switch_to_it((By.ID, "entryIframe"))
        )
        time.sleep(3)

        # 대표 사진 또는 동영상 처리
        photo_elements = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".place_thumb img"))
        )

        # 사진 URL 추출
        photo_url = None
        for photo in photo_elements:
            # 부모 요소에 동영상 관련 클래스가 있는지 확인
            parent_element = photo.find_element(By.XPATH, "..")
            is_video = len(parent_element.find_elements(By.CLASS_NAME, "place_thumb_icon_video")) > 0
            if not is_video:  # 동영상이 아니면 사진 URL 가져오기
                photo_url = photo.get_attribute("src")
                break

        if not photo_url:
            print(f"Store ID {store_id} ({store_name}) 사진이 없습니다.")
            continue

        # 사진 저장
        response = requests.get(photo_url)
        image = Image.open(BytesIO(response.content))
        save_path = os.path.join(save_dir, f"{store_id}.png")
        image.save(save_path, format="PNG")  # PNG 형식으로 저장
        print(f"Store ID {store_id} 사진 저장 완료: {save_path}")

        # iframe 전환 초기화
        driver.switch_to.default_content()

    except Exception as e:
        print(f"Store ID {store_id} ({store_name}) 처리 중 오류 발생: {e}")

# 드라이버 종료
driver.quit()
```
