```python
import os
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

# ChromeDriver 옵션 설정
options = webdriver.ChromeOptions()
options.add_argument("--disable-gpu")
options.add_argument("--window-size=1920,1080")
options.add_experimental_option("excludeSwitches", ["enable-logging", "enable-automation"])
options.add_argument("--disable-blink-features=AutomationControlled")

# ChromeDriver 초기화
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=options)

# 구글 지도에서 크롤링할 리뷰 페이지 링크
google_maps_url = "https://www.google.com/maps/place/%EA%B3%B0%EB%B0%94%EB%A1%9C%EA%B3%B0%ED%83%95/data=!3m1!4b1!4m6!3m5!1s0x357ca151fcaa99d7:0x726e22ef5689292d!8m2!3d37.4910069!4d127.0340103!16s%2Fg%2F11jbd0_165?hl=ko&entry=ttu&g_ep=EgoyMDI0MTExNy4wIKXMDSoASAFQAw%3D%3D"
driver.get(google_maps_url)

# 페이지 로드 대기
time.sleep(5)

# 가게 이름 추출
try:
    store_name_element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//h1[contains(@class, "DUwDvf")]'))
    )
    store_name = store_name_element.text.strip()
    print(f"가게 이름: {store_name}")
except TimeoutException:
    print("가게 이름을 가져올 수 없습니다.")
    driver.quit()
    exit()

# 저장 경로 설정
folder_path = "C:/fintech_service/Fintech_project/google"  # 저장할 폴더 경로를 지정
if not os.path.exists(folder_path):
    os.makedirs(folder_path)
    print(f"디렉토리가 생성되었습니다: {folder_path}")

file_name = f"{store_name.replace(' ', '_')}_reviews_google.csv"
full_path = os.path.join(folder_path, file_name)
print(f"저장 파일 경로: {full_path}")

# 리뷰 섹션의 컨테이너를 찾음
try:
    review_container = WebDriverWait(driver, 30).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, '#QA0Szd > div > div > div.w6VYqd > div:nth-child(2) > div > div.e07Vkf.kA9KIf > div > div > div.m6QErb.DxyBCb.kA9KIf.dS8AEf.XiKgde'))
    )
    
    # 스크롤을 반복적으로 내리기
    max_wait_time = 5  # 최대 대기 시간 시작 (초)
    last_scroll_height = 0
    same_scroll_count = 0  # 스크롤 높이가 변하지 않는 횟수

    while True:
        # 현재 로드된 리뷰 수 확인
        reviews = driver.find_elements(By.CSS_SELECTOR, '.jftiEf.fontBodyMedium')
        previous_count = len(reviews)

        # 스크롤을 아래로 내림
        driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", review_container)
        time.sleep(1)  # 스크롤이 적용되도록 잠시 대기

        # 새로운 리뷰가 로드될 때까지 대기
        try:
            WebDriverWait(driver, max_wait_time).until(
                lambda driver: len(driver.find_elements(By.CSS_SELECTOR, '.jftiEf.fontBodyMedium')) > previous_count
            )
            # 새로운 리뷰가 로드되면 대기 시간을 초기화
            max_wait_time = 5
            same_scroll_count = 0  # 스크롤 상태 초기화
        except TimeoutException:
            # 새로운 리뷰가 로드되지 않으면 스크롤 높이를 비교하여 더 로드할 리뷰가 없는지 확인
            current_scroll_height = review_container.get_attribute("scrollHeight")
            if last_scroll_height == current_scroll_height:
                same_scroll_count += 1
            else:
                last_scroll_height = current_scroll_height
                same_scroll_count = 0
            
            # 동일한 높이가 여러 번 반복되면 종료
            if same_scroll_count > 3:  # 3번 연속 동일하면 로드 완료로 간주
                print("모든 리뷰를 끝까지 로드했습니다.")
                break

            # 스크롤 시간이 길어질 수 있으므로 실패 시 대기 시간을 증가
            max_wait_time += 1

except Exception as e:
    print(f"스크롤 중 오류 발생: {e}")
    driver.quit()
    exit()

# 리뷰 데이터 추출
try:
    reviews = driver.find_elements(By.CSS_SELECTOR, '.jftiEf.fontBodyMedium')
    review_texts = [review.text for review in reviews]
    print(f"총 {len(review_texts)}개의 리뷰를 가져왔습니다.")
    
    # 데이터 저장
    with open(full_path, "w", encoding="utf-8") as file:
        file.write("리뷰 내용\n")
        for review in review_texts:
            file.write(f"{review}\n")
    print(f"리뷰 데이터를 {full_path}에 저장했습니다.")
except Exception as e:
    print(f"리뷰 데이터를 추출하는 중 오류 발생: {e}")

# 브라우저 종료
driver.quit()
```
