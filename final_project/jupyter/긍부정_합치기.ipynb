```python
import pandas as pd
import os

# 원본 파일이 있는 폴더 경로와 저장 폴더 경로 설정
input_directory = "C:/fintech_service/final_project/data/리뷰추출/구글재분석3"
output_directory = "C:/fintech_service/final_project/data/리뷰추출/구글최종전"

# 출력 디렉토리가 없으면 생성
os.makedirs(output_directory, exist_ok=True)

# 디렉토리 내 모든 파일 처리
for file_name in os.listdir(input_directory):
    if file_name.endswith(".csv"):
        input_file_path = os.path.join(input_directory, file_name)
        
        # CSV 파일 로드
        df = pd.read_csv(input_file_path)
        
        # Final_Sentiment 컬럼 생성
        if 'sentiment' in df.columns and 'Sentiment2' in df.columns:
            df['Final_Sentiment'] = df.apply(
                lambda row: row['sentiment'] if row['sentiment'] == '긍정' else row['Sentiment2'],
                axis=1
            )
        
        # 기존 컬럼과 추가 컬럼 순서 재배치
        column_order = ['Date', 'Review', 'Final_Sentiment', '맛', '서비스', '양', 'sentiment', 'Sentiment2']
        df = df[column_order]  # 지정된 순서대로 재배치
        
        # 새로운 파일 이름 생성 및 저장
        output_file_name = f"{os.path.splitext(file_name)[0]}_최종전.csv"
        output_file_path = os.path.join(output_directory, output_file_name)
        df.to_csv(output_file_path, index=False, encoding='utf-8-sig')

        print(f"저장 완료: {output_file_path}")

print("모든 파일 처리가 완료되었습니다!")
```
