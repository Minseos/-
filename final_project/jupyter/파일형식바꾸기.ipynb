```python
import pandas as pd
from dateutil import parser
import re
import os

# 경로 설정
input_folder = "C:/fintech_service/final_project/data/리뷰추출/네이버수정전"
output_folder = "C:/fintech_service/final_project/data/리뷰추출/네이버"
os.makedirs(output_folder, exist_ok=True)

# 날짜 형식을 변환하는 함수
def convert_date(date_str):
    # '2023년 9월 1일 금요일' 형식을 '2023년 9월 1일'로 변경하기 위한 정규식
    date_str = re.sub(r'(\d{4}년 \d{1,2}월 \d{1,2}일) \w+', r'\1', date_str)
    
    # 다양한 형식을 명확하게 시도하여 변환
    formats = [
        "%Y년 %m월 %d일",       # 예시: 2023년 9월 1일
        "%Y/%m/%d",             # 예시: 2023/09/01
        "%Y-%m-%d",             # 예시: 2023-09-01
        "%Y.%m.%d",             # 예시: 2023.09.01
        "%Y%m%d"                # 예시: 20230901
    ]
    
    for fmt in formats:
        try:
            return pd.to_datetime(date_str, format=fmt).strftime('%Y-%m')
        except ValueError:
            continue
    
    # dateutil의 parser로 마지막 시도
    try:
        # 문자열 자동 감지로 변환
        return parser.parse(date_str).strftime('%Y-%m')
    except (ValueError, TypeError):
        # 모든 시도가 실패하면 예외 발생
        raise ValueError(f"Unsupported date format: {date_str}")

# 입력 폴더에서 모든 CSV 파일 처리
for filename in os.listdir(input_folder):
    if filename.endswith(".csv"):
        input_file = os.path.join(input_folder, filename)

        # CSV 파일 불러오기
        df = pd.read_csv(input_file)

        # 'Index' 열 삭제
        if 'Index' in df.columns:
            df = df.drop(columns=['Index'])

        # 'Date' 열이 있는 경우에만 변환 시도
        if 'Date' in df.columns and df['Date'].dtype == 'object':
            try:
                df['Date'] = df['Date'].apply(convert_date)
            except ValueError as e:
                print(f"파일 처리 중 오류 발생: {filename} - {e}")
                continue

        # 변환된 데이터를 새로운 파일로 저장
        output_file = os.path.join(output_folder, filename)
        df.to_csv(output_file, index=False, encoding='utf-8-sig')
        print(f"{filename} 처리 및 저장 완료.")

print("모든 파일이 변환되어 저장되었습니다.")
```

# 카카오 형식 변경 및 더보기 제거

```python
import pandas as pd
from dateutil import parser
import re
import os

# 경로 설정
input_folder = "C:/fintech_service/final_project/data/리뷰추출/카카오수정전"
output_folder = "C:/fintech_service/final_project/data/리뷰추출/카카오"
os.makedirs(output_folder, exist_ok=True)

# 날짜 형식을 변환하는 함수
def convert_date(date_str):
    # '2023년 9월 1일 금요일' 형식을 '2023년 9월 1일'로 변경
    date_str = re.sub(r'(\d{4}년 \d{1,2}월 \d{1,2}일) \w+', r'\1', date_str)
    
    # 다양한 형식을 시도하여 변환
    formats = [
        "%Y년 %m월 %d일",       # 예시: 2023년 9월 1일
        "%Y/%m/%d",             # 예시: 2023/09/01
        "%Y-%m-%d",             # 예시: 2023-09-01
        "%Y.%m.%d",             # 예시: 2023.09.01
        "%Y%m%d"                # 예시: 20230901
    ]
    
    for fmt in formats:
        try:
            return pd.to_datetime(date_str, format=fmt).strftime('%Y-%m')
        except ValueError:
            continue
    
    # dateutil의 parser로 마지막 시도
    try:
        return parser.parse(date_str).strftime('%Y-%m')
    except (ValueError, TypeError):
        raise ValueError(f"Unsupported date format: {date_str}")

# 리뷰 내용에서 '더보기' 제거 함수
def clean_review(text):
    if isinstance(text, str):
        return text.replace("더보기", "").strip()
    return text

# 폴더의 모든 CSV 파일 처리
for file_name in os.listdir(input_folder):
    if file_name.endswith(".csv"):
        input_file = os.path.join(input_folder, file_name)
        
        # CSV 파일 읽기
        df = pd.read_csv(input_file)
        
        # 컬럼명 변경
        df.rename(columns={
            '작성 날짜': 'Date',
            '리뷰 내용': 'Review'
        }, inplace=True)
        
        # 'Date' 열 처리
        if 'Date' in df.columns and df['Date'].dtype == 'object':
            df['Date'] = df['Date'].apply(convert_date)
        
        # 'Review' 열에서 '더보기' 제거
        if 'Review' in df.columns:
            df['Review'] = df['Review'].apply(clean_review)
        
        # '업종 정보' 열 삭제 (존재하는 경우)
        if '업종 정보' in df.columns:
            df = df.drop(columns=['업종 정보'])
        
        # 저장 경로 설정
        output_file = os.path.join(output_folder, file_name)
        
        # 변환된 데이터 저장
        df.to_csv(output_file, index=False, encoding='utf-8-sig')

print("모든 카카오 리뷰 파일이 변환되고 저장되었습니다.")
```

# 구글

```python
import pandas as pd
from datetime import datetime, timedelta
import re
import os

# 현재 날짜 설정
current_date = datetime(2024, 11, 1)  # 기준 날짜를 2024년 11월 1일로 설정

# 입력 및 출력 폴더 경로 설정
input_folder = r'C:\fintech_service\final_project\data\리뷰추출\구글수정전'
output_folder = r'C:\fintech_service\final_project\data\리뷰추출\구글'

# 날짜 형식을 변환하는 함수
def convert_google_date(date_str):
    # "n년 전" 처리
    match_year = re.match(r'(\d+)년 전', date_str)
    if match_year:
        years_ago = int(match_year.group(1))
        target_date = current_date.replace(year=current_date.year - years_ago, month=1, day=1)
        return target_date.strftime('%Y-%m')
    
    # "n달 전" 처리
    match_month = re.match(r'(\d+)달 전', date_str)
    if match_month:
        months_ago = int(match_month.group(1))
        target_date = current_date - timedelta(days=30 * months_ago)
        return target_date.strftime('%Y-%m')
    
    # "n주 전" 처리
    match_week = re.match(r'(\d+)주 전', date_str)
    if match_week:
        weeks_ago = int(match_week.group(1))
        target_date = current_date - timedelta(weeks=weeks_ago)
        return target_date.strftime('%Y-%m')

    # "n일 전" 처리
    match_day = re.match(r'(\d+)일 전', date_str)
    if match_day:
        days_ago = int(match_day.group(1))
        target_date = current_date - timedelta(days=days_ago)
        return target_date.strftime('%Y-%m')

    # 기타 형식은 기본 반환 (변환 불가 시 예외 처리)
    raise ValueError(f"Unsupported date format: {date_str}")

# 폴더 내 모든 CSV 파일 순회
for file_name in os.listdir(input_folder):
    # 파일이 CSV인 경우만 처리
    if file_name.endswith('.csv'):
        input_file_path = os.path.join(input_folder, file_name)
        output_file_path = os.path.join(output_folder, file_name)

        # CSV 파일 불러오기
        df = pd.read_csv(input_file_path)

        # 'Date' 열 변환
        if 'Date' in df.columns and df['Date'].dtype == 'object':
            df['Date'] = df['Date'].apply(convert_google_date)

        # 변환된 데이터를 출력 폴더에 저장
        df.to_csv(output_file_path, index=False, encoding='utf-8-sig')

print("모든 파일의 변환이 완료되었습니다!")
```

# 네이버 재분석 + 네이버 맛_서비스_양

```python
import os
import pandas as pd

# 경로 설정
reanalysis_path = r"C:\fintech_service\final_project\data\리뷰추출\네이버재분석"
reanalysis3_path = r"C:\fintech_service\final_project\data\리뷰추출\네이버재분석3"
final_path = r"C:\fintech_service\final_project\data\리뷰추출\네이버최종"

# 재분석 폴더와 재분석3 폴더의 파일 이름 목록
reanalysis_files = sorted(os.listdir(reanalysis_path))
reanalysis3_files = sorted(os.listdir(reanalysis3_path))

# 병합 작업
for file_name_reanalysis, file_name_reanalysis3 in zip(reanalysis_files, reanalysis3_files):
    reanalysis_file = os.path.join(reanalysis_path, file_name_reanalysis)
    reanalysis3_file = os.path.join(reanalysis3_path, file_name_reanalysis3)
    
    # 파일 읽기
    reanalysis_df = pd.read_csv(reanalysis_file)
    reanalysis3_df = pd.read_csv(reanalysis3_file)
    
    # 필요한 컬럼만 선택 (맛, 서비스, 양)
    columns_to_add = ['맛', '서비스', '양']
    if all(col in reanalysis3_df.columns for col in columns_to_add):
        reanalysis3_selected = reanalysis3_df[columns_to_add]
    else:
        print(f"Warning: {file_name_reanalysis3} 파일에 필요한 컬럼이 없습니다. 건너뜁니다.")
        continue
    
    # 병합 (인덱스를 기준으로 병합)
    merged_df = pd.concat([reanalysis_df, reanalysis3_selected], axis=1)
    
    # 최종 파일 저장
    final_file = os.path.join(final_path, file_name_reanalysis3)
    merged_df.to_csv(final_file, index=False, encoding='utf-8-sig')

print("모든 파일이 성공적으로 병합되어 저장되었습니다!")
```

# 가게 파일 수정

```python
import os
import pandas as pd
import numpy as np  # NaN 처리를 위해 필요

# 디렉토리 경로 설정
google_dir = r'C:\fintech_service\final_project\data\리뷰추출\업종정보\구글'
naver_dir = r'C:\fintech_service\final_project\data\리뷰추출\업종정보\네이버'
kakao_dir = r'C:\fintech_service\final_project\data\리뷰추출\업종정보\카카오'
output_dir = r'C:\fintech_service\final_project\data\리뷰추출\업종정보\병합결과'

# 파일 리스트 가져오기
google_files = os.listdir(google_dir)
naver_files = os.listdir(naver_dir)
kakao_files = os.listdir(kakao_dir)

# 병합 및 저장
for kakao_file in kakao_files:
    kakao_name = kakao_file.split('.')[0]  # 카카오 파일에서 가게 이름 추출
    kakao_path = os.path.join(kakao_dir, kakao_file)
    kakao_data = pd.read_csv(kakao_path)

    # 정확히 일치하는 파일 찾기
    google_match = kakao_file if kakao_file in google_files else None
    naver_match = kakao_file if kakao_file in naver_files else None
    
    # 구글 데이터 읽기
    if google_match:
        google_path = os.path.join(google_dir, google_match)
        google_data = pd.read_csv(google_path)
        # '없음'을 None으로 변환
        google_data['별점'] = google_data['별점'].replace('없음', None)
    else:
        google_data = pd.DataFrame(columns=['가게이름', '별점', '가격대'])  # 빈 데이터
    
    # 네이버 데이터 읽기
    if naver_match:
        naver_path = os.path.join(naver_dir, naver_match)
        naver_data = pd.read_csv(naver_path)
        # '없음'을 None으로 변환
        naver_data['별점'] = naver_data['별점'].replace('없음', None)
    else:
        naver_data = pd.DataFrame(columns=['가게이름', '업종 카테고리', '별점'])  # 빈 데이터
    
    # 카카오 별점에서 0을 NaN으로 변환
    if '별점' in kakao_data.columns:
        kakao_data['별점'] = kakao_data['별점'].apply(lambda x: np.nan if x == 0 else x)
    
    # 전화번호 칸에서 '-'만 있는 경우 None으로 변환
    if '전화번호' in kakao_data.columns:
        kakao_data['전화번호'] = kakao_data['전화번호'].apply(lambda x: None if x.strip() == '-' else x)

    # 가격대 칸에서 특정 값('없음', '₩₩', '₩₩₩', '₩₩₩₩')을 None으로 변환
    if '가격대' in google_data.columns:
        google_data['가격대'] = google_data['가격대'].replace(['없음', '₩₩', '₩₩₩', '₩₩₩₩'], None)

    # 카카오 별점에서 '-'을 None으로 변환
    if '별점' in kakao_data.columns:
        kakao_data['별점'] = kakao_data['별점'].replace('-', None)
    
    # 데이터 병합
    merged_data = pd.DataFrame({
        '가게이름': kakao_name,
        '카테고리': naver_data['업종 카테고리'].iloc[0] if not naver_data.empty else None,
        '주소': kakao_data['주소'].iloc[0] if '주소' in kakao_data.columns else None,
        '전화번호': kakao_data['전화번호'].iloc[0] if '전화번호' in kakao_data.columns else None,
        '영업시간': kakao_data['영업시간'].iloc[0] if '영업시간' in kakao_data.columns else None,
        '가격대': google_data['가격대'].iloc[0] if not google_data.empty else None,
        '네이버 별점': naver_data['별점'].iloc[0] if not naver_data.empty else None,
        '카카오 별점': kakao_data['별점'].iloc[0] if '별점' in kakao_data.columns else None,
        '구글 별점': google_data['별점'].iloc[0] if not google_data.empty else None
    }, index=[0])
    
    # 파일 이름 지정 및 저장
    output_file_name = f"{kakao_name}_정보.csv"
    output_path = os.path.join(output_dir, output_file_name)
    merged_data.to_csv(output_path, index=False, encoding='utf-8-sig')

print("모든 데이터가 병합되고 저장되었습니다!")
```
