```python
import pandas as pd

# 데이터 로드
data = pd.read_csv('./data/updated_negative_text_with_keywords.csv')
review = pd.read_csv('./data/reviews.csv')

# 'review_text' 기준으로 두 데이터 병합
merged_data = pd.merge(data, review[['review_text', 'store_id']], on='review_text', how='left')

# 병합된 데이터 출력
merged_data.head()  # 처음 몇 행만 출력
```

```python
merged_data.to_csv('./data/키워드.csv', index=False)
```
