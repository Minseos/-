```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib
```

```python
from autogluon.tabular import TabularPredictor
```

# 04연속데이터예측_아이오와주_에임스지역_집값예측

```python
data = pd.read_csv('../07_DeepLearning/data/house_train.csv')
data
```

```python
test_data = pd.read_csv('../07_DeepLearning/data/house_test.csv')
test_data
```

```python
# 타겟 변수 설정
target_column = 'SalePrice'
# autogluon 모델 학습
model = TabularPredictor(label=target_column).fit(data)

# 테스트 데이터를 사용해서 예측 수행
predictions = model.predict(test_data)
# 예측 결과 확인
print(predictions)

# 베스트 모델 확인
print("Best model : ", model.get_model_best())

# 모델 성능 평가
if target_column in test_data.columns:
    print('Test score : ', model.evaluate(test_data))
```
