* 1. 데이터 로딩
* 2. 타겟변수 로딩

* 5. 데이터에서 일부 데이터를 테스트 데이터로 추출 - .sample()

* 7. 평가

1. 데이터 로딩

```python
data = pd.read_csv('https://raw.githubusercontent.com/haram4th/ADsP/main/salary2.csv')
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_data, test_data = train_test_split(data, stratify=data['class'], test_size=0.4, random_state=10)
```

# 2. 타겟변수 설정

```python
target_column = 'class'
```

# 3. 제한시간, 검정지표(평가지표, accuracy, RMSE, roc_auc)지정

```python
# 초 단위로 제한시간 지정 300초는 5분,
time_limit = 300
# 성능지표 (accurac, f1, recall, precision, roc_auc_score, mean_squared_error, root_mean_squared_error, r2)
metric = 'accuracy'
```

# 4. 모델정의 TabluarPredictor()

```python
model = TabularPredictor(label=target_column, eval_metric=metric)
```

# 5. 모델 훈련

```python
model.fit(train_data, time_limit=time_limit, presets='medium_quality')
```

# 6. 생성된 모델에 테스트 데이터 넣어 예측하기

```python
pred = model.predict(test_data)
```

# 7. 모델 성능 평가하기

```python
result = model.evaluate(test_data)
result_df = pd.DataFrame([result], index=[0])

# 여러 모델 성능 비교
leader_board = model.leaderboard(test_data)
# 중요 변수 출력
feature_importance = model.feature_importance(test_data)
best_model_name = model.model_best
# best 모델 로딩
best_model = model._trainer.load_model(best_model_name)
best_model_params = best_model.params
```

```python
import gradio as gr
import pandas as pd
from autogluon.tabular import TabularPredictor
from sklearn.model_selection import train_test_split

# automl 함수 정의
def automl(data, target, time=300, metric=None):
    if metric in ('accuracy', 'roc_auc', 'recall', 'precision', 'f1'):
        train_data, test_data = train_test_split(data, stratify=data[target], test_size=0.4, random_state=10)
    else:
        train_data, test_data = train_test_split(data, test_size=0.4, random_state=10)

    model = TabularPredictor(label=target, eval_metric=metric)
    model.fit(train_data, time_limit=time, presets='medium_quality')
    pred = model.predict(test_data)
    
    result = model.evaluate(test_data)
    result_df = pd.DataFrame([result], index=[0])

    leader_board = model.leaderboard(test_data)
    feature_importance = model.feature_importance(test_data)
    best_model_name = model.model_best
    best_model = model._trainer.load_model(best_model_name)
    best_model_params = best_model.params

    return {
        "result_df": result_df.to_html(),
        "leader_board": leader_board.to_html(),
        "feature_importance": feature_importance.to_html(),
        "best_model_name": best_model_name,
        "best_model_params": best_model_params
    }

# 파일을 처리하고 미리보기 및 컬럼 정보를 제공하는 함수
def preview_file(file):
    data = pd.read_csv(file.name)
    preview = data.head()  # 데이터의 상위 5개 행 미리보기
    columns = data.columns.tolist()  # 데이터의 컬럼명 목록
    return preview.to_html(), columns

# automl을 실행하는 함수
def process_file(file, target_column, time_limit, metric):
    data = pd.read_csv(file.name)
    results = automl(data, target_column, time_limit, metric)
    return results['result_df'], results['leader_board'], results['feature_importance'], results['best_model_name'], results['best_model_params']

# Gradio 인터페이스 정의
file_input = gr.File(label="Upload CSV File")
target_input = gr.Textbox(label="Target Column Name")
time_input = gr.Slider(minimum=60, maximum=600, value=300, label="Time Limit (seconds)")
metric_input = gr.Dropdown(choices=["accuracy", "roc_auc", "recall", "precision", "f1", None], label="Evaluation Metric")

# 파일 미리보기와 컬럼 목록 표시용 출력
preview_output = gr.HTML(label="File Preview (Top 5 Rows)")
columns_output = gr.Textbox(label="Available Columns")

# 최종 결과 출력용
result_output = gr.HTML(label="Model Evaluation Results")
leaderboard_output = gr.HTML(label="Leaderboard")
importance_output = gr.HTML(label="Feature Importance")
best_model_name_output = gr.Textbox(label="Best Model Name")
best_model_params_output = gr.Textbox(label="Best Model Parameters")

# Gradio 파일 미리보기 및 컬럼 추출 인터페이스
def preview_and_process_file(file):
    preview_html, columns = preview_file(file)
    columns_str = ", ".join(columns)  # 컬럼을 보기 쉽게 나열
    return preview_html, columns_str

# Gradio 앱 실행
with gr.Blocks() as demo:
    with gr.Row():
        with gr.Column():
            file_input = gr.File(label="Upload CSV File")
            preview_button = gr.Button("Preview File")
            preview_output = gr.HTML()
            columns_output = gr.Textbox(label="Columns")
        
        preview_button.click(preview_and_process_file, inputs=[file_input], outputs=[preview_output, columns_output])

    with gr.Row():
        target_input = gr.Textbox(label="Target Column Name")
        time_input = gr.Slider(minimum=60, maximum=600, value=300, label="Time Limit (seconds)")
        metric_input = gr.Dropdown(choices=["accuracy", "roc_auc", "recall", "precision", "f1", None], label="Evaluation Metric")
        process_button = gr.Button("Run AutoML")
        
    with gr.Row():
        result_output = gr.HTML(label="Model Evaluation Results")
        leaderboard_output = gr.HTML(label="Leaderboard")
        importance_output = gr.HTML(label="Feature Importance")
        best_model_name_output = gr.Textbox(label="Best Model Name")
        best_model_params_output = gr.Textbox(label="Best Model Parameters")
        
        process_button.click(process_file, 
                            inputs=[file_input, target_input, time_input, metric_input], 
                            outputs=[result_output, leaderboard_output, importance_output, best_model_name_output, best_model_params_output])

demo.launch()
```
