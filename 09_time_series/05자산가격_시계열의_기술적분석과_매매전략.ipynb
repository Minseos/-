```python
sym = 'AAPL.O'
data = pd.DataFrame(data[sym]).dropna()
data
```

- 20일간의 애플주식 rolling 지표 구하기

- 최저값: 주어진 데이터 집합에서 가장 작은 값을 의미합니다. 예를 들어, 5개의 데이터가 있을 때 그중에서 가장 작은 수가 최저값입니다.

- 이동평균 (Moving Average): 시간의 흐름에 따라 데이터의 변동을 부드럽게 하기 위해 일정한 구간의 평균을 계산한 값입니다. 예를 들어, 3일 이동평균을 계산할 때는 첫 번째 날부터 세 번째 날까지의 평균, 두 번째 날부터 네 번째 날까지의 평균을 차례대로 계산하는 방식입니다. 이를 통해 데이터의 추세를 파악할 수 있습니다.

- 이동 표준편차 (Moving Standard Deviation): 일정 구간의 데이터에 대해 표준편차(데이터가 평균에서 얼마나 벗어나 있는지를 나타내는 값)를 구하는 방법입니다. 이동평균과 마찬가지로 구간을 이동하며 각 구간의 표준편차를 계산합니다. 이는 데이터의 변동성이나 분산 정도를 파악하는 데 유용합니다.

- 이동 중앙값 (Moving Median): 일정 구간의 데이터에 대해 중앙값(정렬된 데이터의 중간값)을 구하는 방식입니다. 이는 평균 대신 중앙값을 사용해 데이터의 변동을 파악하는 방법으로, 극단값(이상치)에 덜 민감합니다.

- 이동 가중 평균 (Moving Weighted Average): 이동평균과 비슷하지만, 각 데이터에 다른 가중치를 부여하여 평균을 계산하는 방법입니다. 예를 들어, 최근 데이터에 더 높은 가중치를 부여하고 오래된 데이터에는 낮은 가중치를 부여할 수 있습니다. 이렇게 하면 최근 데이터가 더 중요한 영향을 미치게 됩니다.

```python
window=20
```

```python
data['min'] = data[sym].rolling(window).min() # 20일 기간의 최저값
data['mean'] = data[sym].rolling(window).mean() # 20일 간의 이동평균
```

```python
data['std'] = data[sym].rolling(window).std() # 20일 간의 이동 표준편차
data['median'] = data[sym].rolling(window).median() # 20일 간의 이동 중앙값
data['max'] = data[sym].rolling(window).max() # 20일 간의 이동 중앙값
data['ewma'] = data[sym].ewm(halflife=0.5, min_periods=window).mean() # 20일 간의 이동 가중 평균
data
```

```python
data = data.dropna()
```

```python
data
```

- 애플 주가 데이터 시각화 최근데이터 200개만 시각화

```python
ax = data[['min', 'mean', 'max']].iloc[-200:].plot(figsize=(12,6), style=['g--', 'r--', 'g--'], lw=0.8)
data[sym].iloc[-200:].plot(ax=ax, lw=2.0)
plt.title("APPL 20일 Moving Average Price Movement")
plt.show()
```

# 자본시장 시계열의 기술적 분석
* 단기 이동 평균 42일 = 2달
* 장기 이동 평균 252일 = 1년

```python
data['short_ma'] = data[sym].rolling(window=42).mean() # 단기 이동평균 2달
data['long_ma'] = data[sym].rolling(window=252).mean() # 장기 이동평균 1년
data
```

```python
data[[sym, 'short_ma', 'long_ma']].plot(figsize=(12,6))
plt.title("APPL 단순 이동 평균 2달 & 1년")
plt.legend()
plt.show()
```

## 골든 크로스(Golden Cross)
* 단기 이동평균선이 장기 이동 평균선을 아래에서 위로 교차할 때
* 상승장의 시작을 알리는 신호
    * 단기 이동 평균선 : 보통 50일 이동 평균선은 사용
    * 장기 이동 평균선 : 보통 200일 이동 평균선을 사용
## 데드 크로스(Dead Cross)
* 단기 이동 평균선이 장기 이동 평균선을 위에서 아래로 교차할 때
* 하락장을 알리는 신호
    * 단기 이동 평균선 : 보통 50일 이동 평균선을 사용
    * 장기 이동 평균선 : 보통 200일 이동 평균선을 사용

```python
# 골든크로스와 데드 크로스 그래프 같이 그리기
data['positions'] = np.where(data['short_ma'] > data['long_ma'], 1, -1) # 1 = 상승장 매수, -1 = 하락장 매도
ax = data[[sym, 'short_ma', 'long_ma', 'positions']].plot(figsize=(12,6), secondary_y ='positions')
ax.get_legend().set_bbox_to_anchor((0.25, 0.85))
plt.title('AAPL Trading Window based in Technical Analysis')
plt.show()
```
