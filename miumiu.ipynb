```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import pandas as pd
import time

# ChromeDriver 옵션 설정
options = webdriver.ChromeOptions()
options.add_argument("--disable-gpu")
options.add_argument("--window-size=2560,1440")
options.add_experimental_option("excludeSwitches", ["enable-logging", "enable-automation"])
options.add_argument("--disable-blink-features=AutomationControlled")

# ChromeDriver 초기화
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=options)

# URL 접속
url = "https://www.miumiu.com/kr/ko/bags/c/10268KR"
driver.get(url)
time.sleep(3)  # 페이지 로딩 대기

# 쿠키 팝업 처리
try:
    cookie_accept_button = driver.find_element(By.XPATH, '//*[@id="__tealiumGDPRecModal"]/div[2]/div/div/div[1]/div[1]')
    cookie_accept_button.click()
    print("쿠키 팝업 닫기 완료")
    time.sleep(2)  # 팝업 닫힌 후 대기
except Exception as e:
    print(f"쿠키 팝업 처리 중 오류 발생: {e}")

# 모든 상품 로딩 (더보기 버튼 클릭)
while True:
    try:
        load_more_button = driver.find_element(By.CLASS_NAME, "ais-InfiniteHits-loadMore")
        driver.execute_script("arguments[0].click();", load_more_button)
        time.sleep(2)  # 로딩 대기
    except Exception:
        print("모든 상품 로딩 완료 또는 더보기 버튼 없음.")
        break

# 상품 정보 크롤링
product_elements = driver.find_elements(By.CSS_SELECTOR, "li.h-full")
data = []

for product in product_elements:
    try:
        # 상품명
        name = product.find_element(By.CLASS_NAME, "card__title").text
        # 가격
        price = product.find_element(By.CLASS_NAME, "card__price").text
        # 상품 링크
        link = product.find_element(By.TAG_NAME, "a").get_attribute("href")
        # 이미지 URL
        img_url = product.find_element(By.CLASS_NAME, "mobile\\:block").get_attribute("src")
        
        data.append({
            "상품명": name,
            "가격": price,
            "상품 링크": link,
            "이미지 URL": img_url
        })
    except Exception as e:
        print(f"오류 발생: {e}")

# 결과를 DataFrame으로 저장
df = pd.DataFrame(data)

# 엑셀 저장
df.to_excel("miumiu_products.xlsx", index=False)

# 드라이버 종료
driver.quit()
print("크롤링 완료 및 엑셀 저장.")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

# ChromeDriver 옵션 설정
options = webdriver.ChromeOptions()
options.add_argument("--disable-gpu")
options.add_argument("--window-size=2560,1440")
options.add_experimental_option("excludeSwitches", ["enable-logging", "enable-automation"])
options.add_argument("--disable-blink-features=AutomationControlled")

# ChromeDriver 초기화
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=options)

# URL 접속
url = "https://www.miumiu.com/kr/ko/bags/c/10268KR"
driver.get(url)
time.sleep(3)  # 페이지 로딩 대기

# 쿠키 팝업 처리
try:
    cookie_accept_button = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="__tealiumGDPRecModal"]/div[2]/div/div/div[1]/div[1]'))
    )
    cookie_accept_button.click()
    print("쿠키 팝업 닫기 완료")
    time.sleep(2)
except Exception as e:
    print(f"쿠키 팝업 처리 중 오류 발생: {e}")

# 모든 상품 로딩 (스크롤 및 더보기 버튼 반복)
last_height = driver.execute_script("return document.body.scrollHeight")  # 초기 페이지 높이
while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    print("페이지 아래로 스크롤")
    time.sleep(2)

    try:
        load_more_button = driver.find_element(By.CSS_SELECTOR, "#aio > div:nth-child(2) > div > div > div:nth-child(3) > div > div > div.ais-InfiniteHits.plp-results-container > div > a > span")
        load_more_button.click()
        print("더보기 버튼 클릭")
        time.sleep(2)
    except Exception:
        print("더보기 버튼 없음 또는 모든 상품 로드 완료")
    
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:  # 높이가 변하지 않으면 종료
        print("더 이상 로딩할 내용이 없습니다.")
        break
    last_height = new_height

# 상단으로 이동
driver.execute_script("window.scrollTo(0, 0);")
print("페이지 상단으로 이동 완료")
time.sleep(2)

# 스크롤을 천천히 내려 지정된 이미지 URL이 나타날 때까지 진행
while True:
    try:
        # 이미지 URL 요소가 로드되었는지 확인
        img_element = driver.find_element(By.XPATH, '//*[@id="aio"]/div[2]/div/div/div[3]/div/div/div[5]/ol/li[2]/div/article/a/div[1]/div[1]/div/picture/img')
        if img_element.is_displayed():
            print("이미지 URL 요소가 화면에 나타났습니다.")
            break
    except Exception:
        # 스크롤을 조금씩 내림
        driver.execute_script("window.scrollBy(0, 200);")
        print("스크롤을 조금 더 내립니다...")
        time.sleep(1)

# 상품 정보 크롤링
data = []
try:
    product_container = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "#aio > div:nth-child(2) > div > div > div:nth-child(3) > div > div > div.ais-InfiniteHits.plp-results-container > ol"))
    )
    product_elements = product_container.find_elements(By.CLASS_NAME, "h-full")
    
    for product in product_elements:
        try:
            # 상품명
            name = product.find_element(By.CSS_SELECTOR, "div.text-black.pt-4.px-4.pb-8 > h3").text
            # 가격
            try:
                price = product.find_element(By.CSS_SELECTOR, "div.text-black.pt-4.px-4.pb-8 > div:nth-child(2) > p").text
            except Exception:
                price = "가격 정보 없음"
            # 상품 링크
            link = product.find_element(By.TAG_NAME, "a").get_attribute("href")
            # 이미지 URL
            img_url = product.find_element(By.CSS_SELECTOR, "div.card__image-wrapper > div.card__image.card__image--snap > div > picture > img").get_attribute("src")
            
            data.append({
                "상품명": name,
                "가격": price,
                "상품 링크": link,
                "이미지 URL": img_url
            })
        except Exception as e:
            print(f"상품 데이터 크롤링 오류: {e}")
except Exception as e:
    print(f"상품 리스트 크롤링 오류: {e}")

# 결과를 DataFrame으로 저장
df = pd.DataFrame(data)

# 엑셀 저장
output_file = "miumiu_products.xlsx"
df.to_excel(output_file, index=False)
print(f"크롤링 결과가 {output_file} 파일에 저장되었습니다.")

# 드라이버 종료
driver.quit()
print("크롤링 작업 완료.")
```
