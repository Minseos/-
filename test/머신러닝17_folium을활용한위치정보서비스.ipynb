# 시간을 추출해서 출퇴근 시간대 혼잡 역 히트맵 그리기

```python
from datetime import datetime
```

```python
today = datetime.today()
```

```python
# timestamp를 시간 데이터로 변환하기 pd.to_datetime
data['timestamp'] = pd.to_datetime(data['timestamp'])
```

```python
data['timestamp'].dtype
```

```python
# dt.year, month, day, hour, minute, second
# 오전 9시 이전 데이터
morning_data = data[data['timestamp'].dt.hour < 9]
morning_data
```

```python
# 퇴근용 데이터 추출
evening_data = data[data['timestamp'].dt.hour >= 18]
evening_data
```

```python
morning_data.drop('timestamp', axis=1, inplace=True)
evening_data.drop('timestamp', axis=1, inplace=True)
```

```python
morning_station_sum = morning_data.groupby('station_code').sum()
evening_station_sum = evening_data.groupby('station_code').sum()
```

```python
morning_station_sum.head()
```

```python
evening_station_sum.head()
```

```python
morning_joined_data = morning_station_sum.join(station_info)
evening_joined_data = evening_station_sum.join(station_info)
```

```python
morning_joined_data.head(2)
```

```python
evening_joined_data.head(2)
```

```python
# 출근 시간용 지도 만들고 승차 인원 히트맵 지도에 표시하기
# 출근 시간용 서울 지도 만들기
seoul_moring_in = folium.Map(location=[37.55, 126.98], zoom_start=12)
# 히트맵 플러그인을 서울 지도에 추가하기
HeatMap(data = morning_joined_data[['geo.latitude', 'geo.longitude', 'people_in']]).add_to(seoul_moring_in)
seoul_moring_in
```

```python
# 출근 시간용 지도 만들고 하차 인원 히트맵 지도에 표시하기
# 출근 시간용 서울 지도 만들기
seoul_moring_out = folium.Map(location=[37.55, 126.98], zoom_start=12)
# 히트맵 플러그인을 서울 지도에 추가하기
HeatMap(data = morning_joined_data[['geo.latitude', 'geo.longitude', 'people_out']]).add_to(seoul_moring_out)
seoul_moring_out
```

```python
# 퇴근 시간용 지도 만들고 승차 인원 히트맵 지도에 표시하기
# 퇴근 시간용 서울 지도 만들기
seoul_evening_in = folium.Map(location=[37.55, 126.98], zoom_start=12)
# 히트맵 플러그인을 서울 지도에 추가하기
HeatMap(data = evening_joined_data[['geo.latitude', 'geo.longitude', 'people_in']]).add_to(seoul_evening_in)
seoul_evening_in
```

```python
# 퇴근 시간용 지도 만들고 하차 인원 히트맵 지도에 표시하기
# 퇴근 시간용 서울 지도 만들기
seoul_evening_out = folium.Map(location=[37.55, 126.98], zoom_start=12)
# 히트맵 플러그인을 서울 지도에 추가하기
HeatMap(data = evening_joined_data[['geo.latitude', 'geo.longitude', 'people_out']]).add_to(seoul_evening_out)
seoul_evening_out
```
