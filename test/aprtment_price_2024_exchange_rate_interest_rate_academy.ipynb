# 전국 학원 정보

```python
academy = pd.read_csv("../../data/전국부동산실거래가/학원교습소정보_2024년08월31일기준.csv", encoding="cp949")
academy
```

```python
academy.columns
```

```python
academy['시군구'] = academy['시설도로명주소'].str.split()
academy['시군구2'] = academy['시군구'].apply(lambda x: x[:2])
```

```python
# 주소가 빈리스트거나 . 등인 경우 삭제
academy = academy.drop(academy[academy['시설도로명주소'].str.len() < 5].index)
```

```python
def add_name_change(x):
    x = ['서울특별시' if item == '서울' else item for item in x]
    x = ["전북특별자치도" if item == "전라북도" else item for item in x]
    x = ["강원특별자치도" if item == "강원도" else item for item in x] 
    x = ["강원특별자치도" if item == "강원" else item for item in x] 
    x = ["울산광역시" if item == "울산" else item for item in x] 
    x = ["전라남도" if item == "전남" else item for item in x] 
    return x
```

```python
academy['시군구2'] = academy['시군구2'].apply(add_name_change)
```

```python
academy['시군구2'] = academy['시군구2'].apply(lambda x: " ".join(x))
academy
```

```python
for i in academy['시군구2'].unique():
    print(i)
```

```python
academy.loc[academy['시군구2'] == '2층울산광역시 동구', '시군구2'] = '울산광역시 동구'
```

```python
academy.loc[academy['시군구2'] == '5경기도 성남시', '시군구2'] = '경기도 성남시'
```

```python
academy.loc[academy['시군구2'] == '강릉시 주문진읍', '시군구2'] = '강원특별자치도 강릉시'
```

```python
academy.loc[academy['시군구2'] == '서울특별시마포구 토정로17길', '시군구2'] = '서울특별시 마포구'
```

```python
academy.loc[academy['시군구2'] == '서울특별시마포구 새창로', '시군구2'] = '서울특별시 마포구'
```

```python
academy.loc[academy['시군구2'] == '광주광역시?동구?구성로 170', '시군구2'] = '광주광역시 동구'
```
