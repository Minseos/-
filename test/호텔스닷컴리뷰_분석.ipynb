```python
data = pd.read_csv("../data/호텔스닷컴데이터/01_Real_Real_Final_Hotel_KoreanReviewTexts_89942.csv", index_col=0)
data
```

```python
data = data[['rating','description']]
```

```python
data.loc[:,'isgood'] = data['rating'].apply(lambda x: 1 if x >=8 else 0)
```

```python
data = data.reset_index(drop=True)
data
```

```python
data = data[['description', 'isgood']]
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(data, test_size=0.4, stratify=data['isgood'], random_state=777)
```

```python
train.to_csv("../data/호텔스닷컴데이터/hotelscom_review_train.csv", index=False, encoding='utf-8')
```

```python
test.to_csv("../data/호텔스닷컴데이터/hotelscom_review_test.csv", index=False, encoding='utf-8')
```
