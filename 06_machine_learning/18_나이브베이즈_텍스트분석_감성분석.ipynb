```python
def remove_punc(x):
    cleaned =[]
    for i in x:
        if i not in string.punctuation:
            cleaned.append(i)
    #print("cleaned_list : ", cleaned)
    cleaned = "".join(cleaned)
    #print("cleaned_merged : ", cleaned)
    cleaned = cleaned.replace('  ',' ').replace('   ',' ').replace('\n','').replace('\t','').replace('\r','')

    return cleaned
```

# 상투적인 단어 제거하기

```python
for i in data['text'].iloc[0].split():
    print(i)
```

```python
for i in data['text'].iloc[0].split():
    if i not in stopwords.words('english'):
        print(i)
```

```python
stopwords2 = stopwords.words('english')
```

```python
stopwords2.extend(['They','non', 'also'])
```

```python
stopwords2
```

```python
def remove_stopwords(x):
    cleaned =[]
    x = x.lower()
    for i in x.split():
        if i not in stopwords2:
            cleaned.append(i)
    cleaned = " ".join(cleaned)

    return cleaned
```

```python
data['text'] = data['text'].apply(remove_stopwords)
data['text'][:10]
```

```python
X = data[['text']]
y = data['stars']
```

# 자주 등장하는 단어 빈도 분석

```python
whole_words = []
for words in data['text']:
    for word in words.split():
        whole_words.append(word)
```

```python
stopwords2.extend(['place', 'get'])
```

```python
result = []
for i in whole_words:
    if i not in stopwords2:
        result.append(i)
```

```python
from nltk.probability import FreqDist
```

```python
FreqDist(whole_words)
```

# 나이브 베이즈를 이용해서 감성분석하기

## text를 벡터화 하기 CountVectorizer
* 단어의 빈도를 세고 숫자로 변형

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
X_vec = cv.fit_transform(X['text'])
X_vec
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_vec, y, test_size =0.2, random_state=7)
X_valid, X_test, y_valid, y_test = train_test_split(X_valid, y_valid, test_size =0.5, random_state=7)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
mnb = MultinomialNB()
mnb.fit(X_train, y_train)
pred = mnb.predict(X_valid)
print(accuracy_score(y_valid, pred))
print(classification_report(y_valid, pred))
```

```python
# !pip install xgboost
```

```python
from xgboost import XGBClassifier
```

```python
xgb = XGBClassifier(n_estimators=500, max_depth=5, n_jobs= -1, random_state=7)
xgb.fit(X_train, y_train)
pred = xgb.predict(X_valid)
print(accuracy_score(y_valid, pred))
print(classification_report(y_valid, pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=500, max_depth=5, n_jobs= -1, random_state=7)
rfc.fit(X_train, y_train)
pred = rfc.predict(X_valid)
print(accuracy_score(y_valid, pred))
print(classification_report(y_valid, pred))
```
