# 이상값
* 이상값이란 전체 데이터 분포에서 정상 범위에 속하지 않는 데이터
* z-score가  -3 ~ 3 범위를 벗어나는 경우 = 이상값
* z-score 계산하는법
* 평균을 기준으로 평균에서 표준편차 +-3을 넘는 값 = 이상값
* IQR(사분위수) 기준 이상값
   * 하위 이상값 : Q1(25%) - (1.5 * IQR(Q3-Q1))을 넘어가는 값
   * 상위 이상값 : Q3(75%) + (1.5 * IQR(Q3-Q1))을 넘어가는 값

* 단변수 이상치 : 하나의 데이터 분포에서 방생하는 이상치
* 다변수 이상치 : 복수의 연결된 데이터 분포공간에서 발생하는 이상치

* 상자수염 그리고 이상값 탐지하기

```python
data['Fare'].plot(kind = 'box')
```

```python
data['Age'].plot(kind = 'box')
```

```python
data['Fare'].describe()
```

```python
# describe로 구한 Q1, Q3로 이상치 범위 구하기
# 하위 이상값 : Q1(25%) - (1.5 * IQR(Q3-Q1))을 넘어가는 값
# 상위 이상값 : Q3(75%) + (1.5 * IQR(Q3-Q1))을 넘어가는 값
minimun = 7.910400 - (1.5 * (31.000000 - 7.910400))
maximun = 31.000000 + (1.5 * (31.000000 - 7.910400))
print("minimun :", minimun)
print("maximun :", maximun)
```

### 변수간에 상관관계 분석하기
* 상관관계 = -1 0 1의 범위를 갖음
* -1에 가까우면 음의 상관관계가 있음 => 두 변수가 반비례 관계
* 집과 역의 거리가 멀면 월세가 저렴해진다.
* 0은 아무런 관계가 없다.
* +1에 가까우면 양의 상관관계가 있음 => 두 변수가 정비례 관계
* 날씨가 더워지면 에어컨 사용량이 증가한다.

* pearson(피어슨) 상관관계 분석 => 연속형 데이터들 간의 상관관계
* spearman(스피어만) 상관관계 분석 => 서열형, 카테고리형(범주형) 데이터를 포함해서 상관관계 계산

```python
def ages(x):
    for i in data['Age']:
        if 0 < x < 10:
            return('under10')
        elif 10 <= x < 20:
            return('10s')
        elif 20 <= x < 30:
            return('20s')
        elif 30 <= x < 40:
            return('30s')
        elif 40 <= x < 50:
            return('40s')
        elif 50 <= x < 60:
            return('50s')
        elif 60 <= x < 70:
            return('60s')
        elif 70 <= x < 80:
            return('70s')
        elif 80 <= x < 90:
            return('80s')
```

```python
data['Age'] = data['Age'].apply(ages)
```

# 훈련데이터와 테스트 데이터 비율 변화로 성능 튜닝하기

accuracy_score가 가장 높다고 가장 좋은게 아님 f1-score가 높은게 좋은거임

```python
for ts in range(1,10):
    #print(ts/10)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = ts/10, random_state = 8)
    dtc = DecisionTreeClassifier()
    dtc.fit(X_train, y_train)
    pred = dtc.predict(X_test)
    print(ts/10, accuracy_score(y_test, pred))
    print(ts/10, classification_report(y_test, pred))
```
