```python
# 환경 변수 로드
load_dotenv()
client_id = os.getenv("client_id")  # 네이버 api에 접속 가능한 id
client_secret = os.getenv("client_secret")  # 네이버 api에 접속 가능한 pw
start = 1
display = 100
today_date = datetime.now().date()
news_items = []

while True : 
    # API 요청 URL 및 파라미터 설정
    url = "https://openapi.naver.com/v1/search/news.json"
    payload = {'query': '핀테크', 'display': display, 'start': start, 'sort': 'date'}
    headers = {'X-Naver-Client-Id': client_id, 'X-Naver-Client-Secret': client_secret}
     # API 요청
    r = requests.get(url, params=payload, headers=headers)
    
    if r.status_code == 200:
        data = r.json()
        total = int(data['total']) # 475647
        total_page = data['total'] // display + 1
        
        for item in data['items']:
            pub_date_str = item['pubDate']
            pub_date = datetime.strptime(pub_date_str, '%a, %d %b %Y %H:%M:%S %z').date()

            if pub_date == today_date:
                news_items.append(item)

        # 다음 페이지로 이동
        start += display
        if start > total:
            break
        time.sleep(0.5)
        
    else:
        print("API 요청 실패:", r.status_code)
        break

df = pd.DataFrame(news_items)
df
```
