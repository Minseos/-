```python
# 도서상세링크 : 
base_url = "https://www.yes24.com"
links2 = {}

for link2 in soup.select("a.gd_name"):
    href = base_url + link2.get('href')
    links2.setdefault("link", []).append(href)
links2

# soup.select("a[href]")
```

```python
# 저자명
authors = {}
# authPub info_auth 이거는 2개의 클래스를 가지고 있는 것을 의미한다. -> 따라서 중간에 .을 꼭 써야됨
for span in soup.select('span.authPub.info_auth'):
    author = span.find('a').text
    authors.setdefault('author', []).append(author)
authors
```

```python
# 역자명
authors2 = {}
# authPub info_auth 이거는 2개의 클래스를 가지고 있는 것을 의미한다. -> 따라서 중간에 .을 꼭 써야됨
for span in soup.select('span.authPub.info_auth'):
    for a in span.find_all('a'):
        if "역" in span.text:  # '역' 단어가 포함된 부분만 선택
            author2 = a.text.strip()
            authors2.setdefault('author', []).append(author2)

authors2
```

```python
# 출판사명:
pub = {}
for span in soup.select('span.authPub.info_pub'):
    publisher = span.find('a').text
    pub.setdefault("publisher", []).append(publisher)
pub
```

```python
# 출간일:
date = {}
for span in soup.select('span.authPub.info_date'):
    date_text = span.get_text()
    date.setdefault('date',[]).append(date_text)
date
```

```python
# 가격
price = {}
for pri in soup.select('strong.txt_num > em.yes_b'):
    book_price = pri.get_text()
    price.setdefault('Price', []).append(book_price)
price
```

```python
# 적립포인트
point = {}
for po in soup.select("span.yPoint"):
    point_money = po.get_text().replace("포인트적립",'')
    point.setdefault("Point", []).append(point_money)

point
```

```python
# 판매지수
numbers = {}
for num in soup.select("span.saleNum"):
    number = num.get_text().replace("\r\n                                ","").replace('\r\n                            ','')
    numbers.setdefault("Cell_Number", []).append(number)
numbers
```

```python
# 회원리뷰 :
review = {}
for num in soup.select('em.txC_blue'):
    number = num.get_text()
    review.setdefault('review',[]).append(number)
review
```

```python
# 평점
grade = {}
for grades in soup.select("span.rating_grade > em.yes_b"):
    gra = grades.get_text()
    grade.setdefault("Grade", []).append(gra)
grade
```

```python
def text_clean(text):
    return text.replace("\r", "").replace("\n", "").replace("\t", "").strip()
```

```python
book_list_result = {}
for book in book_list:
    kind = book.select_one(".gd_res").string # 종류
    title = book.select_one(".gd_name").string # 책제목
    book_link = "https://www.yes24.com" + book.select_one(".gd_name")['href'] # 도서상세링크
    authors = (book.select_one(".authPub.info_auth").text).split("/")[:-1] if '역' in (book.select_one(".authPub.info_auth").text).split("/")[-1] else book.select_one(".authPub.info_auth").text # 저자명
    trans = (book.select_one(".authPub.info_auth").text).split("/")[-1] if '역' in (book.select_one(".authPub.info_auth").text).split("/")[-1] else "" # 역자명
    publisher = book.select_one(".authPub.info_pub > a").string # 출판사명
    pub_date = book.select_one(".authPub.info_date").string # 출간일
    price = book.select_one(".info_row.info_price > strong > em").string  # 가격
    point = (book.select_one(".info_row.info_price > .yPoint").text).replace("포인트적립",  "") # 포인트
    selling_point = (book.select_one("span.saleNum").text).replace("판매지수",  "").strip() # 판매지수
    n_reviews = book.select_one(".txC_blue").text if book.select_one(".txC_blue") != None else "" # 회원리뷰 수
    rating = book.select_one(".rating_grade > em").text if book.select_one(".rating_grade > em") != None  else "" # 평점
    
    # 결과를 출력합니다.
    #print(kind, title, book_link, authors, trans, publisher, pub_date, price, point, selling_point, n_reviews, rating)
    
    # 값을 리스트로 설정합니다.
    values = [kind, title, book_link, authors, trans, publisher, pub_date, price, point, selling_point, n_reviews, rating]
    keys = ['kind', 'title', 'book_link', 'authors', 'trans', 'publisher', 'pub_date', 'price', 'point', 'selling_point', 'n_reviews', 'rating']
    
    # 결과를 딕셔너리에 저장합니다.
    for key, value in zip(keys, values):
        book_list_result.setdefault(key, []).append(value)

# DataFrame으로 변환하여 CSV 파일로 저장합니다.
final_result = pd.DataFrame(book_list_result)
final_result.to_csv('yes24best_list.csv', encoding='cp949', index=False)
final_result
```
