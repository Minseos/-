```python
import os
from dotenv import load_dotenv
load_dotenv()
import pandas as pd
import requests
import time
from bs4 import BeautifulSoup
```

```python
url = (f"http://openapi.seoul.go.kr:8088/{KEY}/xml/dreamMoney/1/5/")
KEY = os.getenv('KEY')
payload = dict(KEY = 'KEY', TYPE = "xml", SERVICE = "dreamMoney", START_INDEX = 1, END_INDEX = 8)
url = f"http://openapi.seoul.go.kr:8088/{KEY}/xml/dreamMoney/1/5/"
r = requests.get(url, params = payload)
print(r.status_code)
data = r.text
data
```

```python
result = []

soup = BeautifulSoup(data, 'xml')

for item in soup.find_all('row'):
    result2 = {}
    for tags in item:
        result2[tags.name] = tags.text
    result.append(result2)

if result:
    final_df2 = pd.DataFrame(result)
final_df2
```

```python
keys = {}
for item in soup.find_all('row'):
    for tag in item:
        if tag.name == None:
            continue
        else:
            keys.setdefault(tag.name, '')

for key in keys:
    print(soup.find_all(key))
    
```

```python
result = {}
keys = {}
for item in soup.find_all('row'):
    for tag in item:
        if tag.name == None:
            continue
        else:
            keys.setdefault(tag.name, '')

for key in keys:
    for item in soup.find_all(key):
        result.setdefault(item.name, []).append(item.string)
result

if result:
    result_df = pd.DataFrame(result)
result_df


```
