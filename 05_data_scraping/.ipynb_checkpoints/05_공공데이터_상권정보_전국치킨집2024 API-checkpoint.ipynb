```python
store = []
page = 1

while True:
    url = "http://apis.data.go.kr/B553077/api/open/sdsc2/storeListInUpjong"
    servicekey = os.getenv('servicekey')
    payload = dict(servicekey = servicekey, divId = 'indsSclsCd', key = 'I21006', numOfRows = 1000, pageNo = page, type = 'json')
    r = requests.get(url, params = payload)
    #print(r.url)
    
    #print(r.status_code)
    if(r.status_code == 200):
        data = r.json()
        #print(data)
        store.append(data["body"]['items'])
        total_pages = (data["body"]["totalCount"]) // (data['body']['numOfRows']) + 1
    
        if page < total_pages:
            page += 1
        else:
            break
    
    
result = pd.DataFrame()            
for item in store:
    temp = pd.json_normalize(item)
    # temp = pd.json_normalize(item['body']['items'])
    result = pd.concat([result, temp])
    
result.to_csv("chicken.csv", encoding = "cp949")
result
```

```python
# extend를 사용

store = []
page = 1

while True:
    url = "http://apis.data.go.kr/B553077/api/open/sdsc2/storeListInUpjong"
    servicekey = os.getenv('servicekey')
    payload = dict(servicekey = servicekey, divId = 'indsSclsCd', key = 'I21006', numOfRows = 1000, pageNo = page, type = 'json')
    r = requests.get(url, params = payload)
    
    if r.status_code == 200:
        data = r.json()
        store.append(data["body"]['items'])  # extend를 사용하여 리스트를 합침
        total_pages = (data["body"]["totalCount"] // data['body']['numOfRows']) + 1
    
        if page < total_pages:
            page += 1
        else:
            break

# store의 모든 데이터를 포함하는 데이터프레임 생성
result = pd.DataFrame(store)

result.to_csv("chicken.csv2", encoding="cp949", index=False)
print(result)
```

```python
chicken_shops = []
page = 1

while True:
    url = "http://apis.data.go.kr/B553077/api/open/sdsc2/storeListInUpjong"
    payload = dict(servicekey = os.getenv('servicekey'), divId = 'indsSclsCd', key = 'I21006', numOfRows = 1000, pageNo = page, type = 'json')
    r = requests.get(url, params = payload)

    data = r.json()
    chicken_shops.append(data)
    #print(data)
    store.append(data["body"]['items'])
    total_pages = (data["body"]["totalCount"]) // (data['body']['numOfRows']) + 1
    print(f"{page}/{total_pages} 수집중", end = "\r")
    if page < total_pages:
        page += 1
    else:
        break
    time.sleep(0.2)
```

```python
total_chicken_shop_info = pd.DataFrame()
for chicken_shop in chicken_shops:
    temp = pd.json_normalize(chicken_shop['body']['items'])
    total_chicken_shop_info = pd.concat([total_chicken_shop_info,temp]) # 앞에는 빈공간 데이터프레임, 뒤에는 합칠 데이터
total_chicken_shop_info
```
