## 내가 한거

```python
page = 1
all_data =[]
while True:
    url = "https://kind.krx.co.kr/corpgeneral/corpList.do"
    payload = {
        'method': 'searchCorpList',
        'pageIndex': page,
        'currentPageSize': 100,
        'comAbbrv': '',
        'beginIndex': '',
        'orderMode': 3,
        'orderStat': 'D',
        'isurCd': '',
        'repIsuSrtCd': '',
        'searchCodeType': '',
        'marketType': '',
        'searchType': 13,
        'industry': '',
        'fiscalYearEnd': 'all',
        'comAbbrvTmp': '',
        'location': 'all'
    }
    
    r = requests.get(url, params = payload)
   # print(r.status_code)
    #print(r.text)
    
    soup = bs(r.text, 'lxml')
    
    
    titles = [title.get_text() for title in soup.select('td:nth-child(1)')]
    services = [service.get_text() for service in soup.select('td:nth-child(2)')]
    services2 = [services.get_text() for services in soup.select('td:nth-child(3)')]
    dates = [date.get_text() for date in soup.select('td:nth-child(4)')]
    months = [month.get_text() for month in soup.select('td:nth-child(5)')]
    names = [name.get_text() for name in soup.select('td:nth-child(6)')]
    homes = [home.get_text() for home in soup.select('td:nth-child(7)')]
    cities = [city.get_text() for city in soup.select('td:nth-child(8)')]
    
    page_data = list(zip(titles, services, services2, dates, months, names, homes, cities))
    
    if not titles: #page_data
        break
    
    for data in page_data:
        all_data.append(data)
    
    page += 1

columns = ['Title', 'Service', 'Services', 'Date', 'Month', 'Name', 'Home', 'City']

df = pd.DataFrame(all_data, columns = columns)

df.to_csv('company_data.csv', encoding = 'cp949', index = False)
df
```

```python
result = []

# 페이지 번호를 증가시키며 데이터 수집
while True:
    r = requests.get(url, params=payload)
    soup = bs(r.text, 'lxml')

    # 모든 'tr' 요소를 찾기 (데이터가 'tr' 태그에 있다고 가정)
    rows = soup.find_all('tr')
    if not rows:  # 더 이상 데이터가 없으면 종료
        break
    
    for row in rows:
        columns = row.find_all('td')
        if columns:
            row_data = [column.get_text(strip=True) for column in columns]
            result.append(row_data)
    
    print(f"Page {payload['pageIndex']} processed")
    
    # 페이지 인덱스 증가
    payload['pageIndex'] += 1

# DataFrame으로 변환
if result:
    columns = ['회사명', '업종', '주요제품', '상장일', '결산월', '대표자명', '홈페이지', '지역']
    final_df = pd.DataFrame(result, columns=columns[:len(result[0])])
# DataFrame 출력
final_df
```

```python
page = 1
page_size = 100
final_result_df = pd.DataFrame()
while True:
    url = "https://kind.krx.co.kr/corpgeneral/corpList.do"
    payload = dict(method='searchCorpList',pageIndex=page, currentPageSize=page_size,orderMode=3,orderStat='D',searchType=13, fiscalYearEnd='all', location='all')
    r = requests.get(url, params=payload)
    print(r.status_code, end="\r")
    soup = bs(r.text, 'lxml')
    
    total_items = int(soup.select_one(".info.type-00 > em").text.replace(",", ""))
    total_pages = total_items // page_size + 1
    
    print(f"{page}/{total_pages} 수집중", end="\r")
    keys = soup.select_one("table.list.type-00.tmt30")['summary'].split(", ") 
    
    result = {}
    for tr in soup.select('tr'):
        for idx, (key, td) in enumerate(zip(keys, tr.select('td'))):
            if idx == 0:
                kinds = [img['alt'].strip() for img in td.select('img')]   # 1번째 증권 종류, 회사이름
                kind = ", ".join(kinds)
                code = td.select_one("a")['onclick'].split("'")[1] # 종목코드 추출
                result.setdefault('증권종류', []).append(kind)
                result.setdefault(key, []).append(td.text)
                result.setdefault('종목코드', []).append(code+"0") # 종목코드 추가
            elif idx == 6:
                home_link = td.select_one('a')['href'] if td.string == None else ""  # 6번째 링크 찾기
                result.setdefault(key, []).append(home_link)
            else:
                result.setdefault(key, []).append(td.text)
    result_df = pd.DataFrame(result)
    final_result_df = pd.concat([final_result_df, result_df])
        
    if page < total_pages:
        page += 1
        time.sleep(5)
    else:
        break

display(final_result_df)
final_result_df.to_csv(f"상장기업정보_{today}기준.csv", encoding='utf-8', index=False)  # 수집일 기준으로 csv 파일로 저장
```
