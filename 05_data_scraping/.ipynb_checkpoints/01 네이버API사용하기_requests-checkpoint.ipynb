### while True로 전체 데이터 가져오기

```python
# 강사님이 푼 거

book_lists = []
start = 1
page = 1
query = input("검색어를 입력하세요 : ")
while True :
    client_id = "hu54ROaBRw1_p6uW63lp"            # 네이버 api에 접속 가능한 id
    client_secret = "rWNfODtDBd"                  # 네이버 api에 접속 가능한 pw
    encText = urllib.parse.quote(query)
    #print(encText)
    url = f"https://openapi.naver.com/v1/search/book.json?query={encText}&display=10&start={start}" # JSON 결과
    #print(url)
    try:
        request = urllib.request.Request(url)
        request.add_header("X-Naver-Client-Id",client_id)
        request.add_header("X-Naver-Client-Secret",client_secret)
        response = urllib.request.urlopen(request)
        rescode = response.getcode()
        
        if(rescode==200):
            response_body = response.read()
            data = json.loads(response_body.decode('utf-8'))
            book_lists.append(json.loads(response_body.decode('utf-8')))
            total_pages = data['total'] // 10 + 1
    
            if page > total_pages:
                break;
        else:
            print("Error Code:" + rescode)
        if page < total_pages:
            page += 1
            if start != 991:
                start += 10
            elif start == 991:
                start += 9
            print(f"{page:03d}/ {total_pages:03d} start : {start} 추출중", end ="\r")
        else: 
            break
        time.sleep(1)
    except Exception as e:
        print(e)
        break
    
        
    print(len(book_lists))
    result = pd.DataFrame()
    for book_list in book_lists:
        temp = pd.json_normalize(book_list['items'])
        result = pd.concat([result, temp])
    print(result)
    result.to_csv(f"naver_book_api_fintech_{query}result2.csv", encoding = "cp949")
```

3. 함수화

```python
def select():
    topic = input("서비스를 입력하세요.(블로그, 책, 뉴스, 전문자료)")
    if topic == "블로그":
        topic = "blog"
    elif topic == "책":
        topic = "books"
    elif topic == "뉴스":
        topic = "news"
    elif topic == "전문자료":
        topic = "doc"
    return topic

def search():
    search = input("검색어를 입력하세요 : ")
    return search

service = select()
query = search()

book_lists = []
start = 1
page = 1
#query = input("검색어를 입력하세요 : ")
while True :
    client_id = "hu54ROaBRw1_p6uW63lp"            # 네이버 api에 접속 가능한 id
    client_secret = "rWNfODtDBd"                  # 네이버 api에 접속 가능한 pw
    encText = urllib.parse.quote(query)
    #print(encText)
    url = f"https://openapi.naver.com/v1/search/{service}.json?query={encText}&display=10&start={start}" # JSON 결과
    #print(url)
    try:
        request = urllib.request.Request(url)
        request.add_header("X-Naver-Client-Id",client_id)
        request.add_header("X-Naver-Client-Secret",client_secret)
        response = urllib.request.urlopen(request)
        rescode = response.getcode()
        
        if(rescode==200):
            response_body = response.read()
            data = json.loads(response_body.decode('utf-8'))
            book_lists.append(json.loads(response_body.decode('utf-8')))
            total_pages = data['total'] // 10 + 1
    
            if page > total_pages:
                break;
        else:
            print("Error Code:" + rescode)
        if page < total_pages:
            page += 1
            if start != 991:
                start += 10
            elif start == 991:
                start += 9
            print(f"{page:03d}/ {total_pages:03d} start : {start} 추출중", end ="\r")
        else: 
            break
        time.sleep(1)
    except Exception as e:
        print(e)
        break
    
        
    print(len(book_lists))
    result = pd.DataFrame()
    for book_list in book_lists:
        temp = pd.json_normalize(book_list['items'])
        result = pd.concat([result, temp])
    print(result)
    result.to_csv(f"naver_{service}_api_fintech_{query}_result.csv", encoding = "cp949")
```
